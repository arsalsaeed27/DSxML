```python
# Cleaning the dataset
data.isnull().sum()
```

```python
data = data.drop_duplicates()
data
```

```python
# Since ['Gross'] has too many nulls in it, ill drop it
data = data.drop(['Gross'], axis =1)
# or data = data.drop(columns = "Gross")
```

```python
data
```

```python
data["GENRE"] = data["GENRE"].str.lstrip("\n")
data["ONE-LINE"] = data["ONE-LINE"].str.lstrip("\n")
data["ONE-LINE"] = data["ONE-LINE"].str.rstrip("\n")
data["STARS"] = data["STARS"].str.lstrip("\n")
```

```python
data
```

```python
data['ONE-LINE']
```

```python
data['YEAR']
# this is weird, i only want one year
```

```python
# extracting first 4digit year only to get out of confuison
data['YEAR'] = data['YEAR'].str[1:5]
```

```python
data[data['YEAR'].isnull()]
# these are the movies whose year we dont know. since year is a critical column, ill drop all these rows because theyre of no use to me
```

```python
data = data.dropna(subset=['YEAR'])
data
```

```python
# since there are still duplicate movies in it. ill remove the rows which have same names of movies
data = data.drop_duplicates(subset=['MOVIES'])
data
```

```python
data.isnull().sum()
# now we can see our data is massively cleared from nulls, especially the important column of year
```

```python
# replacing NaNs in the column
data['RATING'] = data['RATING'].fillna(data['RATING'].median())
data['RunTime'] = data['RunTime'].fillna(data['RunTime'].median())
data['VOTES'] = data['VOTES'].fillna('Unknown')
data
```

```python
data.isnull().sum()
```

```python
# since nan genre rows are very less, ill drop it cuz it wont make much of a difference
data = data.dropna(subset=['GENRE'])
data
```

```python
data[data['ONE-LINE'] == "Add a Plot"]
```

```python
# since there are 281 movies with no plot given, ill just change it to not available
data['ONE-LINE'] = data['ONE-LINE'].replace('Add a Plot', 'Not Available')
```

```python
data
```

```python
data['STARS'] = data['STARS'].str.replace('\n',' ')
data
```

```python
# data[data['STARS'] == '']
data['STARS'] = data['STARS'].replace('','Not Available')
data
```

```python
data['Director'] = data['STARS'].str.split('|').str[0]
# data['Director']
data['Director'] = data['Director'].where(data['Director'].str.contains('Director:'), 'Not Available')
data['Director'] = data['Director'].str.replace('Director:', '' ).str.replace('Directors:' ,'').str.strip()
data
```

```python
data['Stars'] = data['STARS'].str.split('|').str[-1]
# data['Stars']
```

```python
data['Stars'] = data['Stars'].where(data['Stars'].str.contains('Stars:') | data['Stars'].str.contains('Star:'), 'Not Available')
data['Stars'] =  data['Stars'].str.replace('Stars:', '' ).str.replace('Star:', '').str.strip()
data
```

```python
data = data.drop(['STARS'], axis =1)
```

```python
data
data = data[['MOVIES', 'YEAR', 'GENRE', 'RATING', 'ONE-LINE', 'Director', 'Stars', 'VOTES', 'RunTime']]
data
```
