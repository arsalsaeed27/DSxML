```python
delivery = pd.read_csv("/content/deliveries.csv")
delivery
```

```python
#isin() function
death_over = delivery[delivery['over']>15]
death_over
```

```python
allbatsman= death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x= allbatsman>200
batsman_list = allbatsman[x].sort_values(ascending=False).index.tolist()
```

```python
final = death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head()
```
