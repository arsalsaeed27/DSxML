```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit Sharma Vs Virat Kohli (IPL)')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(['Virat Kohli','Rohit Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linewidth=2)
plt.title('Rohit Sharma Vs Virat Kohli (IPL)')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(['Virat Kohli','Rohit Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linewidth=2,marker='o',markersize=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linewidth=2,marker='o',markersize=4)
plt.title('Rohit Sharma Vs Virat Kohli (IPL)')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(['Virat Kohli','Rohit Sharma'])
```

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
children = [10,40,20,10,30]
colors = ['red red red red red','green green green green green','blue blue blueblueblue ','yellow','pink']
plt.bar(colors,children)
plt.xticks(rotation=90)
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom = df['2016'] + df['2017'],label='2015')
plt.legend()
```
