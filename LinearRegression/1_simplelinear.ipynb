```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/placement.csv')
```

```python
df.head()
```

this is a random dataset having final gpa and salary of students. so i'll train my model using the followigng data and then i'll use it to see if its reliable or not.

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA  ')
plt.ylabel('PACKAGE')
```

```python
x= df.iloc[:,0:1].values
y= df.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

Alright so my model is trained now, i'll now just start predcting tha salaries with respect to final gpa

```python
x_test
```

```python
y_test
```

```python
# x_test[0] = 8.58 cgpa and the salary is 4.1 lakh
lr.predict(x_test[0].reshape(1,1))
# so according to my model, 8,58 cgpa will land a 3.89 lakh salary job, but it is not precise
```

```python
lr.predict(x)
```

```python
plt.plot(x,lr.predict(x))
plt.xlabel('CGPA')
plt.ylabel('SALARY')
```

The straight line shows as CGPA increases, the predicted Salary increases linearly.

```python
plt.scatter(x, y, color='blue', label='Actual Data')   # actual data points
plt.plot(x, lr.predict(x), color='red', label='Regression Line')  # predicted line
plt.xlabel('CGPA')
plt.ylabel('SALARY')
plt.legend()
```

This showw actual data points as blue dots and the best fit regression line in red which my trained model used to estimate the salaries.
this means that according to my model, if cgpa is 6, the salary should be 2.5.

```python
# x_test[3] is 6.22. and its predicted salary is 2.57. just like i said above.
lr.predict(x_test[3].reshape(1,1))
```
